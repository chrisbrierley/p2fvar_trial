### Mapping

Mapping requires the following inputs in the NCL script:
- experiment names 
- variable names
- filename

In this case, these parameters are being written directly in the NCL file when running the following cell.

```python
# press Enter + Shift to run this cell
exptname = "C20-Reanalysis"           #experiment name
varname  = "nino34_spacomp_tas_djf1"  #variable name

# --------------------------------------------------------------------------
filename = "PMIP4_output/"+exptname+"_"+varname+".png"
```

```python
%%bash -s "$exptname" "$varname"
ncl_command=`echo ncl \'exptname=\"$1\"\' \'varname=\"$2\"\' PMIP4_scripts/plot_expt_map.ncl`
echo running $ncl_command
echo 
eval $ncl_command
```
